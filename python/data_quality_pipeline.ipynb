```python
import pandas as pd
```

```python
customers = pd.read_csv("Customers.csv")
inventory = pd.read_csv("Inventory.csv")
order_items = pd.read_csv("Order_Items.csv")
orders = pd.read_csv("Orders.csv")
products = pd.read_csv("Products.csv") 
stores = pd.read_csv("Stores.csv")
suppliers = pd.read_csv("Suppliers.csv")

print("Files Loaded Succesfully")
```

```python
# VALIDATION CHECKS
report = {}

# Null checks
report["null_customer_in_orders"] = orders["customer_id"].isnull().sum()
report["null_product_in_orderitems"] = order_items["product_id"].isnull().sum()

# Duplicate checks
report["duplicate_orders"] = orders["order_id"].duplicated().sum()

# Negative checks
report["negative_quantity"] = (order_items["quantity"] <= 0).sum()
report["negative_unit_price"] = (order_items["unit_price"] <= 0).sum()

# Selling price anomaly
report["selling_less_than_cost"] = (products["selling_price"] < products["cost_price"]).sum()

df_report = pd.DataFrame(report.items(), columns=["Check", "Issue_Count"])
df_report
```

```python
# Saving Report
df_report.to_csv(r"C:\Retail_Data\data_quality_report.csv", index=False)

print("Data quality report saved successfully.")
```
